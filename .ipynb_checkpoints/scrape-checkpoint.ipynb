````python
def get_requests_page(page_num):
    url = 'https://buscador.plataformadetransparencia.org.mx/web/guest/buscadornacional?p_p_id=Buscador_WAR_Buscador&p_p_lifecycle=2&p_p_state=normal&p_p_mode=view&p_p_resource_id=informacion&p_p_cacheability=cacheLevelPage'
    params = {"contenido":"covid","tipoColeccion":2,"pagina":page_num,"listaTipoColeccion":["5"],"filtrosSeleccionados":[],"contruirFiltro":False,"filtrosDescartados":[]}
    headers = {
        'Content-Type':'application/json',
        'Origin': 'https://buscador.plataformadetransparencia.org.mx',
        'Cookie': '__utma=266489167.345104395.1621487680.1621487680.1621487680.1; __utmc=266489167; __utmz=266489167.1621487680.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); __utmb=266489167.1.10.1621487680; COOKIE_SUPPORT=true; GUEST_LANGUAGE_ID=es_ES; JSESSIONID=Q8svTUPQbf7W4aMK1It3xpAgP0nINdwsBFFUTocl.aquiles; _ga=GA1.3.345104395.1621487680; _gid=GA1.3.1054066756.1621487698; LFR_SESSION_STATE_20104=1621488778825',
    }
    return requests.post(url, data=json.dumps(params), headers=headers).json()
````

````python
output_data = []
````

````python
import time
````

````python
for i in range(0,2929):
    print(str(i))
    try:
        data = get_requests_page(i)
    except:
        data = {}
    if len(list(data.keys()))>0:
        output_data.append(data)
    time.sleep(0.05)
````
